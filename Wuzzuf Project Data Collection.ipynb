```python
import csv
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

## make an input for user to search what is the job he is need

```python
url = input("Enter The job that you want to find ")
s = url.split()
if len(s) == 2:
    result = f'https://wuzzuf.net/search/jobs/?a=navbl&q={s[0]}%20{s[1]}&start=0'
elif len(s) == 3:
    result = f'https://wuzzuf.net/search/jobs/?a=navbl&q={s[0]}%20{s[1]}%20{s[2]}&start=0'
```

```python
result
```

```python
response = requests.get(result)
response
```

## Get Soup

```python
page = 0  # counter to go to the next page 
data = []  # for store data
while True: # make a while loop to make a program run in all the page in the jobs that you search 
    if len(s) == 2:
        result = f'https://wuzzuf.net/search/jobs/?a=navbl&q={s[0]}%20{s[1]}&start={page}'
    elif len(s) == 3:
        result = f'https://wuzzuf.net/search/jobs/?a=navbl&q={s[0]}%20{s[1]}%20{s[2]}&start={page}'
    get_result = requests.get(result)  # make a response for the job that user want to see
    soup = BeautifulSoup(get_result.content,'html.parser') # make an instance from Beautifulsoup , make the structure of the page in html 
    all_jobs = soup.find_all('div',attrs={'class':'css-pkv5jc'})  # the page of the job
    job_num = soup.find('li', attrs = {'class' : 'css-8neukt'}).text # get number of the jobs 
    for j in all_jobs:  # make a loop to collect data form the pages and store the data 
        job_name = j.find('h2').get_text() # get name of the job
        job_link = j.find('h2').find('a').get('href') # get the link of the job
        job_description = j.find('div',attrs={'class':'css-y4udm8'}).get_text() # get the description of the job
        job_location = j.find('span').get_text() # get the location of the job
        company_name = j.find('a', attrs = {'target' : "_blank", 'class' : 'css-17s97q8'}).get_text() # get company name
        # List to collect the data
        d = [job_name,job_link,job_description,job_location,company_name]
        data.append(d) # append the data in the d list 
    if job_num[:-3].find(job_num[-3:]) == -1:  # make a condition to count the page and when the numebr of the job is finished stop the while loop
        page +=1
    else:
        print("Finished")
        break
    print('Page Number :',page)  # print number of the job
```

```python
len(data)
```

## Make csv File to store the data 

```python
with open('JobsOfWuzzuf.csv',"w",newline="",encoding='UTF-8') as f:
    w = csv.writer(f)
    header = ["Job Name ","Job Link","Job Description","Job Location","Company Name"]
    w.writerow(header)
    w.writerows(data)
```
